```python
import matplotlib.pyplot as plt 
import numpy as np 

fig,ax=plt.subplots()

x= np.linspace(0, 20, 1000)
ax.plot(x,np.cos(x))
ax.axis('equal')

ax.annotate('local maxium',xy=(6.28,1),xytext=(10,4),
           arrowprops=dict(facecolor='b',shrink=0.04))

ax.annotate('local minimum',xy=(5*np.pi,-1),xytext=(2,-6),
           arrowprops=dict(arrowstyle='->',connectionstyle="angle3,angleA=0,angleB=90"));
```

# Annotate & legend

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.arange(0,10)
y=x^2
z=x^2

plt.plot(x,y)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.arange(0,10)
y=x^2
z=x^3

plt.plot(x,y,marker='o')
plt.plot(x,z)
plt.plot(y,z)
plt.xlabel('x')
plt.ylabel('y')
plt.title('MY GRAPH')

#annotation
plt.annotate(xy=[2,1],text="entry",arrowprops=dict(facecolor='b'))
plt.annotate(xy=[4,6],text="entry",arrowprops=dict(facecolor='b'))
plt.annotate(xy=[8,10.5],text="entry",arrowprops=dict(facecolor='b'))


plt.legend(["Race1","race2","race3"],loc="lower right")
plt.show()

```

# pie plot 

pie chart is a circular statistical plot that can display only one series of data

```python
import numpy as np
import matplotlib.pyplot as plt

#creating dataset
Branch=["computer","electrical","civil","mechanical","automobile"]
data=[50,33,40,30,11]
explode =(0.3,0,0,0,0)


fig=plt.figure(figsize=(10,7))
plt.pie(data,labels= Branch,explode=explode,autopct='%1.1f%%',shadow=True)
plt.show()
```

# 3D  Ploting

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d


fig=plt.figure()

ax=plt.axes(projection='3d')

z=np.linspace(0,1,100)
x=z*np.sin(25*z)
y=z*np.cos(25*z)

ax.plot3D(x,y,z,'r')
ax.set_title('3D ploting')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Taking Co-ordinates of X and Y
x = [5,10,15,20,25,30,35,40,45,50]
y = [1,4,3,2,7,6,9,8,10,5]

plt.plot(x,y,"b")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.tick_params(axis="x", labelsize = 12, labelrotation = 20, labelcolor = "red")
plt.tick_params(axis="y", labelsize = 12, labelrotation = 20, labelcolor = "blue")

#Save This Chart
plt.savefig("TestChart.pdf",format='pdf')

plt.show()
```
